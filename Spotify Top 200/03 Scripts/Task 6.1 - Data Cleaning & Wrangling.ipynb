### BR, IN, & US all need to have null values removed & 'Position' reassigned as int, before we can merge all country data into since df

```python
print("BR:", BR.shape,
     "IN:", IN.shape,
     "US:", US.shape)
```

```python
BR.dropna(subset=['Position'], inplace = True)
IN.dropna(subset=['Position'], inplace = True)
US.dropna(subset=['Position'], inplace = True)
```

```python
print("BR:", BR.shape,
     "IN:", IN.shape,
     "US:", US.shape)
```

```python
BR['Position'].isnull().values.any()
```

```python
IN['Position'].isnull().values.any()
```

```python
US['Position'].isnull().values.any()
```

```python
BR['Position'] = BR['Position'].astype('int64')
IN['Position'] = IN['Position'].astype('int64')
US['Position'] = US['Position'].astype('int64')
```
